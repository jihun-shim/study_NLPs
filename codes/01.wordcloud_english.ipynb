## 빈도수에 의한 자연어 처리

- 이 코드는 단어의 빈도를 기반으로 텍스트 데이터를 처리하고, WordCloud를 사용해 단어를 시각화하는 과정을 보여줍니다. 아래에 각 단계별로 초보자가 이해하기 쉽도록 설명

### 데이터 전처리

```python
it_sentences = [
    "Artificial intelligence technology is increasingly being applied to various sectors of society",
    "Cloud computing is making data storage and processing more efficient and is revolutionizing the IT infrastructure of organizations.",
    "Machine learning and deep learning algorithms are revolutionizing data analysis and predictive modeling.",
    "Blockchain technology is helping to improve the transparency and security of financial transactions.",
    "Cybersecurity is emerging as a critical issue in the digital age and requires constant attention and improvement.",
    # "Agile methodologies in software development enable organizations to respond effectively to rapid market changes.",
    # "Data science and analytics play an important role in the business decision-making process and make organizations more competitive.",
    # "Internet of Things (IoT) technologies are transforming everyday life into a more convenient and connected place.",
    # "Virtual Reality (VR) and Augmented Reality (AR) technologies are providing new experiences in entertainment, education, healthcare, and more.",
    # "Python is a widely used language for a variety of programming tasks, including data science, web development, and automated scripting."
]

# '''
# 설명: 분석에 사용할 IT 관련 문장들을 준비합니다.
# '''
```

```python
text = ''.join(it_sentences)
type(text), len(text)

# '''
# 2. 문자열 변환 및 분리

# 1) ''.join(it_sentences)는 문장 리스트를 하나의 문자열로 만듭니다.
# 2) .split(sep=' ')는 공백(' ')을 기준으로 단어를 나눕니다.
# '''
```

```python
text_split = text.split(sep = ' ') # 띄어쓰기로 분리하기
type(text_split), len(text_split), text_split[:4] # 단어 출력 몇개인지

# '''
# type(text_split): 분리된 데이터의 타입(리스트)을 확인합니다.
# len(text_split): 총 몇 개의 단어가 있는지 확인합니다.
# text_split[:4]: 리스트의 첫 4개 단어를 출력하여 내용 확인.
# '''
```

### 단어(명사, 품사, 대명사 등등) 사전

```python
# 단어마다 빈도수
from collections import Counter
counter_text_split = Counter(text_split) # 중복된 단어 출력시 사용
type(counter_text_split), counter_text_split.most_common()[:10] # get list ,list 얻기
# word cloud 에서 불용어 처리 필요 (and, is, of, the 등)

# '''
# 3. 단어 빈도수 계산

# Counter는 각 단어의 등장 횟수를 세는 데 사용됩니다.

# 결과:
# 가장 많이 등장한 단어 10개를 출력합니다.
# 예: [('is', 6), ('and', 4), ('the', 3), ...]
# '''
```

### word cloud

```python
from wordcloud import WordCloud # is, the, an 등등 없애는 모듈 즉 불용어 없애는 모듈

# '''
# WordCloud란?

# 텍스트 데이터를 시각화하는 기법으로, 단어의 빈도수에 따라 크기를 다르게 표현합니다.
# '''
```

```python
wordCloud = WordCloud(background_color='white')

# '''
# background_color='white': WordCloud의 배경색을 흰색으로 설정합니다.
# '''
```

```python
wordCloud.generate_from_text(text)

# '''
# generate_from_text(text): 텍스트 데이터를 기반으로 WordCloud를 만듭니다.
# '''
```

```python
wordCloud.words_ # 확률로 담아 보여줌
```

```python
import matplotlib.pyplot as plt
plt.imshow(wordCloud)
plt.axis('off') # 그래프 좌표 없앰
plt.show()

# '''
# imshow()는 생성된 WordCloud 이미지를 시각화합니다.
# axis('off')를 통해 그래프의 축을 숨겨 깔끔한 출력을 만듭니다.
# '''
```
